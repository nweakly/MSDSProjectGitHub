```python
#Drop duplicate records for the same-day observations and reset the index
accidents_ann2.drop_duplicates(inplace=True)
accidents_ann2.reset_index(inplace=True, drop=True)
```

```python
#merge observations for each day 
accidents_anndaily3 = pd.merge(accidents_anndaily, accidents_ann2, how='left', left_on=accidents_anndaily.index, right_on='Date') 
```

```python
#check results
accidents_anndaily3.head()
```

```python
#rename a column
accidents_anndaily3.rename(columns={'Date_x':'NumberOfAccidents'}, inplace=True)
# Drop 'Date_y' column
accidents_anndaily3.drop('Date_y', axis=1, inplace=True)
#Drop 'Date' column
accidents_anndaily3.drop('Date', axis=1, inplace=True)
#check results
accidents_anndaily3.info()
```

```python
accidents_anndaily3.head()
```

## Use ANN (Artificial Neural Network) to forecast the daily number of accidents

```python
#additional imports
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```python
#set up target variable and independent variables
X=accidents_anndaily3.drop('NumberOfAccidents', axis=1) #independent variables
y=accidents_anndaily3['NumberOfAccidents'] #target variable
```

```python
#Split the dataset into training and test part
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Create an instance of the model
mlp=MLPRegressor(hidden_layer_sizes=(20,20,20,20), max_iter=500)
```

```python
# fit training data
mlp.fit(X_train, y_train)
```

```python
#generate predictions
predictions=mlp.predict(X_test)
```

```python
#Evaluate prediction results
#R2 score
mlp.score(X_test, y_test) 
```

```python
#Evaluate results
import sklearn.metrics
```

```python
#Explained variance
sklearn.metrics.explained_variance_score(y_test, predictions)
```

```python
# Mean absolute error
sklearn.metrics.mean_absolute_error(y_test, predictions)
```

```python
# Mean squared error
sklearn.metrics.mean_squared_error(y_test, predictions)
```

```python
# Median absolute error
sklearn.metrics.median_absolute_error(y_test, predictions)
```

```python
#R2, coefficient of determination, describes how well future observations are likely to be predicted [-1, 1]
sklearn.metrics.r2_score(y_test, predictions) 
```


Overall, the model showed a relatively low predictive power, as according to the coefficient of determination it is likely to predict 37.7% of the future variations of the dependent variable.
Adjusting the model parameters (number of hidden layers, neurons, learning rate, the maximum number of iterations) did have any significant influence on the model's predictive power.
