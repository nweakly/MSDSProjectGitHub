### Rush Hour Traffic

```python
# extract time of the accident
accidents_all3['Time'] = accidents_all3['FIRST_OCCURENCE_DATE'].dt.time
```

```python
# check results
accidents_all3.head()
```

```python
# Function to check if accident time falls within traditional rush hours  07:00:00 - 09:00:00 and 16:00:00 - 18:00:00
# Note: this function checks only times and does not discriminate between weekdays, weekends and holidays

def rush_hour(x):
    """ checks if timestamp falls within rush hour periods (07:00:00 - 09:00:00 and 16:00:00 - 18:00:00)
    """
    start1 = datetime.time(7, 0,0)
    end1 = datetime.time(9,0,0)
    start2 = datetime.time(16,0,0)
    end2 = datetime.time(18,0,0)
    if x >= start1:
        if x <= end1:
            return 1
        elif x >= start2:
            if x <= end2:
                return 1       
    return 0
```

```python
# Create the "RushHour' column with values: 1  = accident happened during rush hours, 0 - not during rush hours
# Let's consider times from 07:00:00 to 09:00:00 and from 16:00:00 to 18:00:00 On 24-hour clock to be rush hour traffic 
accidents_all3['RushHour'] = accidents_all3['Time'].apply(lambda x: rush_hour(x))
# Drop the 'Time' column
accidents_all3.drop('Time', axis=1, inplace=True)
```



Use four periods in the day depending on natural light conditions:

- astronomical twilight (astroT) = [from astroT Start to mauticalT start] AND [from nauticalT end to astroT end]
- nautical twilight (nauticalT) = [from nauticalT start to civilT start] AND [from CivilT end to nauticalT end]
- civil twilight= [from civil twilight start to sunrise] AND [from sunset to civil twilight end] 
- daytime (day) = [from sunrise to sunset]

If an observation does not have '1' in either of twilight periods or daytime period columns, it means that this accident happened during the night time conditions: night = [from to 00:00:00 to Astronomical twilight start] AND [ after astronomical twilight end to 23:59:59]



```python
# Functions to check if accident happened during daylight or any of the special 
# twilight periods

def daylight(x):
    """ returns 1 if timestamp falls within daytime
    """
    if x['FIRST_OCCURENCE_DATE'] >= x['Sunrise']:
        if x['FIRST_OCCURENCE_DATE'] < x['Sunset']:
            return 1
    return 0
    
    
def astro_twilight(x):
    """ checks if timestamp falls within astronomical twighlight periods
    """
    if x['FIRST_OCCURENCE_DATE'] >= x['AstroT_Start']:
        if x['FIRST_OCCURENCE_DATE'] < x['NauticalT_Start']:
            return 1
        elif x['FIRST_OCCURENCE_DATE'] >= x['NauticalT_End']:
            if x['FIRST_OCCURENCE_DATE'] < x['AstroT_End']:
                return 1       
    return 0

def nautical_twilight(x):
    """ checks if timestamp falls within nautical twighlight periods
    """
    if x['FIRST_OCCURENCE_DATE'] >= x['NauticalT_Start']:
        if x['FIRST_OCCURENCE_DATE'] < x['CivilT_Start']:
            return 1
        elif x['FIRST_OCCURENCE_DATE'] >= x['CivilT_End']:
            if x['FIRST_OCCURENCE_DATE'] < x['NauticalT_End']:
                return 1       
    return 0

def civil_twilight(x):
    """ checks if timestamp falls within civil twighlight periods
    """
    if x['FIRST_OCCURENCE_DATE'] >= x['CivilT_Start']:
        if x['FIRST_OCCURENCE_DATE'] < x['Sunrise']:
            return 1
        elif x['FIRST_OCCURENCE_DATE'] >= x['Sunset']:
            if x['FIRST_OCCURENCE_DATE'] < x['CivilT_End']:
                return 1       
    return 0
```

```python
# Create the "Daytime' column with values: 1  = accident happened during daytime, 0 - not during daytime 
accidents_all4['Daytime'] = accidents_all4.apply(daylight, axis=1)
```

```python
# Create the "AstroTwilight' column with values: 1  = accident happened during astronimical twilight time, 0 - not during astronomical twitlight time 
accidents_all4['AstroTwilight'] = accidents_all4.apply(astro_twilight, axis=1)
```

```python
# Create the "NauticalTwilight' column with values: 1  = accident happened during nautical twilight time, 0 - not during nautical twitlight time 
accidents_all4['NauticalTwilight'] = accidents_all4.apply(nautical_twilight, axis=1)
```

```python
# Create the "CivilTwilight' column with values: 1  = accident happened during civil twilight time, 0 - not during civil twitlight time 
accidents_all4['CivilTwilight'] = accidents_all4.apply(civil_twilight, axis=1)
```
